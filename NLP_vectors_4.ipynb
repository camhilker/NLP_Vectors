#### Plotting 1 desert sample

Here I took 1 desert sample and tokenized it over each reading frame for k-mers of length 20. These were then replaced with 10-dim vectors that were assigned to the training set of k-mers if a match was found. The vectors for each sequence were then averaged, ending up with one 10-dim vector to represent each sequence with a match to the plasmid k-mers. There were 2,663 sequences with a match out of 5,670,854.

```python
import pandas as pd
import numpy as np
import csv

from Bio import Entrez
from time import sleep

import matplotlib.pyplot as plt
import seaborn as sns

import gensim
import pickle

import os
import gzip

from sklearn.cluster import KMeans, DBSCAN, MeanShift, estimate_bandwidth, AgglomerativeClustering
from sklearn.decomposition import PCA
```

```python
X = pd.read_csv('desert_sample.csv')
```

```python
pca = PCA(n_components=2)
pca.fit(X)
pca_X = pca.transform(X)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(pca_X[:, 0], pca_X[:, 1])
```
